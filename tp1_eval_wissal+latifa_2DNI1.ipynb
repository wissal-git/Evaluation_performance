## Exemple 1 :

<h3>
<span style="color:indigo "> Distribution uniforme continue sur l'intervalle [0,1] </span>
    </h3>
    
 - une variable alatoire uniforme discrte signifir que les n valeurs discretres x1,x2,...xn de la variable alatoire x ont chacune la mme probabilits d'appirition  <span style="color:red ">P(X=xi)=1/N</span>.
    

- dans cette exemple on vient de developper deux fonctions calculons la moyenne et la variance empirique, qui vont tres utiliser dans les autres exemple, et on les appliquer sur une liste de taille N de valeurs alatoires. 
=> On obtenant les resultats on les compare avec les mesures thoriques :

**moyenne thorique : E(x) =1/2**

**variance thorique : V(x)=1/12**

<h3>
<span style="color: orange ;"> - Etape 1 : gnration d'un tableau de N variable alatoire </span>
    </h3>

```python

import random
def Les_variables(N):
    tab=[]
    for i in range (N) :
        tab.append(random.uniform(0,1))
    return tab
```

<h3>
<span style="color: orange ;">- Etape 2 : cration des fonction: moyenne empirique et variance empirique 
    </span>
    </h3>

```python
def Moyenne_n(tab, N):
    s1=0
    for i in tab:
        s1=s1+i
    return s1/N
```

```python
def variance_n(tab, N, moyenne):
    s2=0
    for i in tab:
        s2=s2+(i-moyenne)**2
    return s2/N
```

<h3>
<span style="color: orange;">- Etape 3 :figure reprsentatif des valeurs trouves pour diffrent N :
    </span>
    </h3>

 - calcul de moyenne pour chaque N
 - calcul de la variance pour chaque N;   avec N=nombre des variable alatoire
 - affichage du calcul

```python
import matplotlib.pyplot as plt
N=[10, 100, 1000, 10000, 100000]
Les_Moyennes=[]
Les_Variances=[]
for i in N :
    tab=Les_variables(i)
    Les_Moyennes.append(Moyenne_n(tab, i))
    Les_Variances.append(variance_n(tab, i, Moyenne_n(tab, i)))

plt.plot(N, Les_Moyennes, 'g')

plt.plot(N, Les_Variances, 'r')
plt.show()
for i in range (len(N)):
    print('les valeurs alatoire pour n=',N[i],'leurs moyenne =',Les_Moyennes[i],'leurs variance =',Les_Variances[i])
```
