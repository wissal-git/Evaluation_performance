## Exemple 1 :

<h3>
<span style="color:indigo "> Distribution uniforme continue sur l'intervalle [0,1] </span>
    </h3>
    
 - une variable aléatoire uniforme discréte signifir que les n valeurs discretres x1,x2,...xn de la variable aléatoire x ont chacune la méme probabilités d'appirition  <span style="color:red ">P(X=xi)=1/N</span>.
    

- dans cette exemple on vient de developper deux fonctions calculons la moyenne et la variance empirique, qui vont étres utiliser dans les autres exemple, et on les appliquer sur une liste de taille N de valeurs aléatoires. 
=> On obtenant les resultats on les compare avec les mesures théoriques :

**moyenne théorique : E(x) =1/2**

**variance théorique : V(x)=1/12**

<h3>
<span style="color: orange ;"> - Etape 1 : génération d'un tableau de N variable aléatoire </span>
    </h3>

```python

import random
def Les_variables(N):
    tab=[]
    for i in range (N) :
        tab.append(random.uniform(0,1))
    return tab
```

<h3>
<span style="color: orange ;">- Etape 2 : création des fonction: moyenne empirique et variance empirique 
    </span>
    </h3>

```python
def Moyenne_n(tab, N):
    s1=0
    for i in tab:
        s1=s1+i
    return s1/N
```

```python
def variance_n(tab, N, moyenne):
    s2=0
    for i in tab:
        s2=s2+(i-moyenne)**2
    return s2/N
```

<h3>
<span style="color: orange;">- Etape 3 :figure représentatif des valeurs trouvées pour différent N :
    </span>
    </h3>

 - calcul de moyenne pour chaque N
 - calcul de la variance pour chaque N;   avec N=nombre des variable aléatoire
 - affichage du calcul

```python
import matplotlib.pyplot as plt
N=[10, 100, 1000, 10000, 100000]
Les_Moyennes=[]
Les_Variances=[]
for i in N :
    tab=Les_variables(i)
    Les_Moyennes.append(Moyenne_n(tab, i))
    Les_Variances.append(variance_n(tab, i, Moyenne_n(tab, i)))

plt.plot(N, Les_Moyennes, 'g')

plt.plot(N, Les_Variances, 'r')
plt.show()
for i in range (len(N)):
    print('les valeurs aléatoire pour n=',N[i],'leurs moyenne =',Les_Moyennes[i],'leurs variance =',Les_Variances[i])
```

```python
import math
import random 

def Les_variables_4(N):
    tab4=[]
    proba=[]
    for _ in range(N):
        x=random.uniform(0,1)
        proba.append(x)
#print(lst)    
 
             
    for j in proba:
        if j<0.3:
            tab4.append(0)
        elif ((j>=0.3) and (j<0.5)):
            tab4.append(2)
        else:
            tab4.append(6)           
    return tab4
```
